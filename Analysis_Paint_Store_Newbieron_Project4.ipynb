```python
# Import require library for creating charts.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Now import the csv file and save into a variable.
pst = pd.read_csv('Paint Store Details.csv')
pst = pd.DataFrame(pst)
pst.head()
```

```python
# We have converted all 'NaN' to 'NA' but when upload the csv file then see all changed to 'NaN' so convert once again.
# we can apply 'np.nan' with 'replace' function to do that.
pst = pst.replace(np.nan,'NA',regex=True)
```

```python
# checking the data set.
pst.head()
```

```python
# For safty reason we should make a copy of data set.
mst=pst.copy(deep=True)
```

```python
# Just To check the data.
mst.head()
```

```python
pst['Exists Of Url'] = mst['Web Url'].apply(lambda x: x.split(":")[0])
```

```python
pst['Exists Of Url'] = pst['Exists Of Url'].str.replace("NA","No").replace("https","Yes").replace("http","Yes")
```

```python
pst.head(10)
```

###**Compare with each states in India where they do have website or not.**

```python
# This is the plot size configaration.
plt.figure(figsize=(10,10))

# Now we can use plot style but it's not mandatory.
sns.set_style("whitegrid")

# In order to sort the data in decending order.
mf_sh = pst.sort_values(by=['State/Territory'],ascending=False)

# We used seaborn to making the diagram, Y axis is define 'State/Territory' and hue or we can say legend is define 'Exists Of Url' columns and mention the data set.
mi = sns.countplot(y='State/Territory',hue='Exists Of Url', data=pst)

# We created this forloop for mention the count on each bar.
for i in mi.containers:
  mi.bar_label(i, fmt='%.0f', label_type='edge')

plt.xlabel("Count Of Website Url") # This is title of x axis.
plt.ylabel("States") # This is title of Y axis.
plt.title("State wise count of website url for each paint shop in India") # This is Title of Diagram.
```

Upon generating the reports, it becomes apparent that several states lack their own dedicated websites. Uttar Pradesh emerges as one of the top states with the highest number of missing websites, closely followed by Maharashtra. Conversely, Tamil Nadu and Madhya Pradesh exhibit the highest number of websites, with Andhra Pradesh securing the second-highest position. It is imperative to prioritize states without websites and encourage them to establish an online presence to foster business growth and development.
